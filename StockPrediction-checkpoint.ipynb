```python
data = pd.read_csv("CementSectorData.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
symbols= data['Symbol'].unique().tolist() #saving all unique company names in a list
print (symbols)
```

```python
len(symbols)
```

```python
companies = data.groupby('Symbol')#grouping data with respect to companies
```

```python
companies.first()  #will return first entry of every company
```

```python
def Grouping(li):
    for i in li:
        j=companies.get_group(i) #grouping all data of each company and yielding the data 
        yield j
```

```python
sectors=symbols.copy()
print(sectors)
```

```python
for value in Grouping(symbols):      #checking functon yielding correct values or not
    print(value)
```

```python
i=0
for value in Grouping(symbols):
    sectors[i]=value
    if(i<len(sectors)):
        i=i+1
# saving dataframes at each location of the list yield by the function grouping    
```

```python
for i in range(0,len(symbols)):
    sectors[i].to_csv(symbols[i]+'.csv',index=False, encoding='utf8')
#saving data of different companies to different files     
```
